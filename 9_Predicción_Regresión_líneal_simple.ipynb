```python
# Cargar datos
df = pd.read_excel("/content/Ajuste modelo de regresion lineal_LHV (1).xlsx")
df.head(2)
```

```python
# Seleccionar variables
variable_x = "ENVERGADURA"
variable_y = "ESTATURA"
```

```python
# Importar librerías necesarias
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Supongamos que df es tu DataFrame
variable_x = 'ESTATURA'
variable_y = 'ENVERGADURA'

# Generar análisis
modelo = LinearRegression()
modelo.fit(df[[variable_x]], df[variable_y])

# Imprimir ecuación de la recta
print('Ecuación de la recta: y =', round(modelo.coef_[0], 3), 'x +', round(modelo.intercept_, 3))

# Imprimir coeficiente de correlación
print('Coeficiente de correlación:', round(np.corrcoef(df[variable_x], df[variable_y])[0, 1], 3))

# Imprimir coeficiente de determinación
print('Coeficiente de determinación:', round(r2_score(df[variable_y], modelo.predict(df[[variable_x]])), 3))

# Gráfica con intervalo de confianza
sns.regplot(x=df[variable_x], y=df[variable_y], ci=95, line_kws={"color": "red"}, scatter_kws={"color": "blue"})
plt.title('Regresión lineal simple con intervalo de confianza al 82,8%')
plt.show()
```

```python
# Generar PREDICCIÓN
dato_predictor = 1.7

# Crear DataFrame con el dato predictor
ESTATURA = pd.DataFrame([dato_predictor], columns=[variable_x])

# Realizar la predicción
prediccion_envergadura = modelo.predict(ESTATURA)

# Imprimir la predicción
print('La predicción para una estatura de', dato_predictor, 'es una envergadura de', round(prediccion_envergadura[0], 3))
```
